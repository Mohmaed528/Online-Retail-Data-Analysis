#  Visualization

```python
data['Month'] = data['InvoiceDate'].dt.to_period('M')
```

```python
quantity_monthly = data.groupby('Month')['Quantity'].sum().reset_index()
quantity_monthly['Month'] = quantity_monthly['Month'].astype(str)

sns.set_style('whitegrid') 
plt.figure(figsize=(18,10))

sns.lineplot(
    data=quantity_monthly,
    x='Month',
    y='Quantity',
    marker='o',            
    linewidth=2.5,        
    markersize=10,         
    color='royalblue'      
)

# Ø¹Ù†Ø§ÙˆÙŠÙ† ÙˆÙˆØ¶Ø¹ Ø§Ù„Ø®Ø·ÙˆØ·
plt.title("Monthly Quantities", fontsize=22, fontweight='bold')
plt.xlabel('Month', fontsize=16)
plt.ylabel('Quantity', fontsize=16)
plt.xticks(rotation=45)     

plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()     
plt.show()
```

```python
quantity_monthly = (
    data.groupby('Month')['Quantity']
    .sum()
    .reset_index()
    .sort_values(by='Quantity', ascending=False)
)
quantity_monthly['Month'] = quantity_monthly['Month'].astype(str)

sns.set_style('whitegrid')
plt.figure(figsize=(18,10))

ax = sns.barplot(
    data=quantity_monthly,
    x='Month',
    y='Quantity',
    palette='RdBu_r',        
    edgecolor='black'       
)

plt.title("Monthly Quantities", fontsize=22, fontweight='bold')
plt.xlabel('Month', fontsize=18)
plt.ylabel('Quantity', fontsize=18)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

for bars in ax.containers:
    ax.bar_label(bars, fontsize=16, padding=3)  

plt.tight_layout()
plt.show()
```

```python
quantity_pos = data[data['Quantity'] > 0]
quantity_describ = (
    quantity_pos.groupby('Description')['Quantity']
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

plt.figure(figsize=(16,10))
ax = sns.barplot(
    data=quantity_describ,
    x='Description',
    y='Quantity',
    palette='coolwarm',    
    edgecolor='black'      
)

plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)

for bars in ax.containers:
    ax.bar_label(bars, fontsize=14, padding=3, alpha=0.8)

plt.tight_layout()
plt.title('Top 10 Products by Quantity Sold', fontsize=20, fontweight='bold')
plt.xlabel('Product Description', fontsize=16)
plt.ylabel('Quantity Sold', fontsize=16)

plt.show()
```
