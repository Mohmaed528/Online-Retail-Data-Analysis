#  SUBSCRIBERS

```python
data.info()
```

```python
data['published_timestamp'] = data['published_timestamp'].str.split('T').str.get(0)
```

```python
data['published_timestamp'] = data['published_timestamp'].replace("3 hours",'2016-06-21')
```

```python
data['published_timestamp'] = pd.to_datetime(data['published_timestamp'],format='%Y-%m-%d')
```

```python
data['Year'] = data['published_timestamp'].dt.year
data['Month'] = data['published_timestamp'].dt.month
data['Day'] = data['published_timestamp'].dt.day
```

```python
data.head()
```

```python
data.groupby('Year')['num_subscribers'].sum()
```

```python
plt.figure(figsize=(15,8))
plt.style.use('ggplot')

ax=sns.barplot(data=data,x='Year',y='num_subscribers',hue='subject',ci=None,width=.8)
plt.xlabel("Years",fontsize = 15)
plt.ylabel('Num Of Subscribers',fontsize = 15)
for bars in ax.containers:
    ax.bar_label(bars,fmt='%.0f',fontsize=9)

plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.style.use('ggplot')

plt.xlabel("Years",fontsize = 15)
plt.ylabel('Num Of Subscribers',fontsize = 15)

sns.lineplot(data=data,x='Year',y='num_subscribers',hue='subject',ci=None,linewidth=2.5)
plt.show()
```

# LEVELS

```python
data['level'].value_counts()
```

```python
plt.figure(figsize=(10,8))
ax = sns.countplot(x=data['level'])
plt.xlabel('Levels',fontweight="bold")
for bars in ax.containers:
    ax.bar_label(bars,color="black")
```

```python
plt.figure(figsize=(10,8))

ax = sns.barplot(data=data,x='level',y='num_subscribers',hue='subject',ci=None)
plt.xlabel("Years",fontsize = 15)
plt.ylabel('Num Of Subscribers',fontsize = 15)
for bars in ax.containers:
    ax.bar_label(bars,fmt="%.0f")
plt.show()
```

```python
plt.figure(figsize=(12,8))

num_subjects_level=data.groupby('subject')['level'].value_counts().sort_values(ascending=False).plot(kind='bar',color='tab:blue',fontsize=12)

plt.show()
```

```python

def level_charts_per_y(y):
    plt.figure(figsize = (20,10))
    plt.xlabel("Levels",fontsize=15,fontweight='bold',labelpad=20)
    plt.ylabel(f"{y}",fontsize=15,fontweight='bold',labelpad=20)
    plt.title(f'plot of {y} per level')
    ax=sns.barplot(data=data,x='level',y=y,hue='subject',ci=None,width=.7)
    for bars in ax.containers:
        ax.bar_label(bars,fontsize=12,fmt='{:.1f}')
    
    plt.show()
    

    plt.figure(figsize = (10,5))
    data.groupby('level')[y].sum().plot(kind='pie',shadow=True)
    plt.title(f'pie chart of {y} per level')
    plt.show()
```

```python
level_charts_per_y('num_reviews')
```

```python
data['price'] = data['price'].replace('Free|TRUE','0',regex=True)
```

```python
data['price'] = data['price'].astype(float)
```

```python
data['Profit'] = data['price'] * data['num_subscribers']
```

```python
plt.figure(figsize = (10,8))

ax=sns.barplot(data=data,x='Year',y='Profit',ci=None,estimator=sum)
for bars in ax.containers:
    ax.bar_label(bars,fmt='{:.0f}')
plt.show()
```
